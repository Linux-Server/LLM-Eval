```python
import os
from deepeval.openai import OpenAI
from deepeval.models import GPTModel
from dotenv import load_dotenv
load_dotenv()

llm = GPTModel(model="gpt-5", base_url=os.getenv("LIGHTAI_BASE_URL"), api_key=os.getenv("LIGHTAI_API_KEY"))

llm.generate("hi")
```

```python
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.models import AnthropicModel
from deepeval.test_case import LLMTestCase
from deepeval import assert_test

answer_metric = AnswerRelevancyMetric(model=llm)

test_case = LLMTestCase(
    input="Who is half blood prince?",
    actual_output="Severus Snape"

)
assert_test(test_case=test_case, metrics=[answer_metric])
```

```python
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.test_case import LLMTestCase
from deepeval import evaluate
from deepeval.dataset import EvaluationDataset, Golden
from deepeval.models import AnthropicModel

model = AnthropicModel(model="claude-sonnet-4-5")

answer_metric = AnswerRelevancyMetric(model=model, async_mode=False)

```

```python
from datasets import load_dataset

hugg_data = load_dataset("Sandhya2002/harry_potter_dataset", split="train")
```

```python
hugg_data = hugg_data.remove_columns(['id', 'status', 'inserted_at', 'updated_at', '_server_id', 'question_quality.responses', 'question_quality.responses.users', 'question_quality.responses.status'])
```

```python
hugg_data[0]
```

```python
hugg_goldens = []

from deepeval import evaluate
from deepeval.dataset import EvaluationDataset, Golden

for data in hugg_data:
    hugg_goldens.append(Golden(input=data["Question"], actual_output=data["Answer"]))
```

```python
dataset = EvaluationDataset(goldens=hugg_goldens)
```

```python
for goldens in dataset.goldens:
    dataset.add_test_case(
        LLMTestCase(
            input=goldens.input,
            actual_output=goldens.actual_output
        )
    )
```

```python
dummy_test = dataset.test_cases[:3]
```

```python
from deepeval.evaluate import AsyncConfig
evaluate(test_cases=dummy_test, metrics=[answer_metric])
```
